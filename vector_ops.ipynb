```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
v1=np.array([3,4])
v2=np.array([1,2])

suma=v1+v2
resta=v1-v2
producto_escalar=np.dot(v1,v2)

print(suma,resta,producto_escalar)

```

```python
fig,ax=plt.subplots()

ax.quiver(0,0,v1[0],v1[1],angles='xy',scale_units='xy',scale=1,color='r',label='Vector 1')
ax.quiver(0,0,v2[0],v2[1],angles='xy',scale_units='xy',scale=1,color='b',label='Vector 2')

ax.quiver(0,0,suma[0],suma[1],angles='xy',scale_units='xy',scale=1,color='g',label='Vector Suma')
ax.quiver(0,0,resta[0],resta[1],angles='xy',scale_units='xy',scale=1,color='m',label='Vector Resta')

ax.set_xlim(-1,7)
ax.set_ylim(-1,7)
ax.axhline(0,color='white',linewidth=0.5)
ax.axvline(0,color='white',linewidth=0.5)
ax.grid(True,which='both')

plt.legend()
plt.title('Operaciones con vectores')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
v1=np.array([3,4,5])
v2=np.array([1,2,3])
```
